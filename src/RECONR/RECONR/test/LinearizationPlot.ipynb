This jupyter notebook is used to visualize (not verify) the linearization capabilities in RECONR. The linearziation is rather course because we used a relative tolerance of 0.1. 

```python
import matplotlib.pyplot as pyplot
```

```python
def plotDifference(refX, refY, linX, linY, MT):
    fig, ax = pyplot.subplots()
    ax.set_xscale("log")
    ax.set_yscale("log")
    ax.plot(linX, linY, "-o", color="blue", label="linearized")
    ax.plot(refX, refY, "o", color="red", markerfacecolor="white", label="initial")
    ax.set_title("MT={}".format(MT))
    ax.legend()
    return ax
```

## MT=1

```python
refX = [1e-05,2e-05,750000,1.9e+07,1.95e+07,2e+07]
refY = [1,11.829,3.34739e-05,2.75176e-05,2.7313e-05,2.71079e-05]
linX = [1e-05,2e-05,2e-05,750000,750000,1.9e+07,1.9e+07,1.95e+07,1.95e+07,2e+07,2e+07]
linY = [1,1,11.829,11.829,3.34739e-05,3.34739e-05,2.75176e-05,2.75176e-05,2.7313e-05,2.7313e-05,2.71079e-05]

len(linX)
```

```python
ax = plotDifference(refX, refY, linX, linY, 1)
```
